```python
def func_with_io():
    for i in range(int(1e5)):
        print('/', end='')
        x = 1
```

```python
def func():
    for i in range(int(1e5)):
        x = 1
```

```python
def func_file_io():
    for i in range(int(1e5)):
        fp = open('test.txt', 'w')
        x = 1
        fp.write(str(x))
        fp.close()
```

```python
%load_ext line_profiler

%lprun -u 1e-3 -f func func()
```

```python
%lprun -u 1e-3 -f func_with_io func_with_io()
```

```python
%lprun -u 1e-3 -f func_file_io func_file_io()
```
